Importing Required Libraries

```python
# Importing necessary libraries for data processing, visualization, and model building
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
```

Loading the Dataset

```python
# Load the dataset (replace with your file path if needed)
df = pd.read_csv('creditcard_2023.csv')

# Display basic information about the dataset
df.info()

# Display first few rows of the dataset
df.head()
```

Checking for Duplicate Rows

```python
# Check if there are any duplicate rows in the dataset
duplicate_rows = df[df.duplicated()]

# Print the number of duplicate rows found
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")

# Remove duplicate rows if any
df_cleaned = df.drop_duplicates()

# Confirm that duplicates have been removed
print(f"Dataset shape after removing duplicates: {df_cleaned.shape}")
```

Visualizing Class Distribution

```python
# Check the distribution of the target variable 'Class' (0 = Not Fraud, 1 = Fraud)
sns.countplot(x='Class', data=df_cleaned, palette='coolwarm')
plt.title('Class Distribution (0: Not Fraud, 1: Fraud)')
plt.show()
```

Identifying Features with High Skewness


```python
# Check skewness of numerical features
skewness = df_cleaned.select_dtypes(include=[np.number]).skew()

# Display features with skewness greater than 1 or less than -1
print("Features with high skewness (|skew| > 1):")
print(skewness[abs(skewness) > 1])
```

Log Transformation to Address Skewness

```python
# Apply log transformation to reduce skewness in the 'Amount' feature
df_cleaned['Log_Amount'] = np.log1p(df_cleaned['Amount'])

# Check the distribution of the transformed 'Amount' feature
sns.histplot(df_cleaned['Log_Amount'], bins=50, kde=True)
plt.title('Log-Transformed Transaction Amount')
plt.show()
```

Feature Selection Using Correlation Matrix

```python
# Calculate the correlation matrix to identify highly correlated features
correlation_matrix = df_cleaned.corr()

# Plot the correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()
```

Scaling Features

```python
# Standardize the features (important for anomaly detection algorithms)
scaler = StandardScaler()

# Select features for scaling (excluding the 'Class' target variable)
features = df_cleaned.drop(columns=['Class', 'Log_Amount'])
scaled_features = scaler.fit_transform(features)

# Convert scaled features into a DataFrame for better readability
scaled_features_df = pd.DataFrame(scaled_features, columns=features.columns)

# Display the first few rows of the scaled data
scaled_features_df.head()
```

Visualizing the Scaled Data Using PCA

```python
# Use PCA for dimensionality reduction to visualize the scaled data in 2D
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features_df)

# Create a DataFrame with the PCA components
pca_df = pd.DataFrame(data=pca_components, columns=['PCA1', 'PCA2'])
pca_df['Class'] = df_cleaned['Class'].reset_index(drop=True)

# Visualize the 2D PCA projection
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Class', data=pca_df, palette='coolwarm')
plt.title('PCA of Scaled Data')
plt.show()
```

Checking for Outliers in Features

```python
# Visualizing potential outliers using boxplots for key features like 'Amount' and 'Time'
plt.figure(figsize=(12, 6))

# Boxplot for 'Amount' feature
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Amount'])
plt.title('Boxplot for Transaction Amount')

# Boxplot for 'Time' feature
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Class'])
plt.title('Boxplot for Transaction Time')

plt.tight_layout()
plt.show()
```

Creating an Anomaly Detection Pipeline

```python
# Creating a function to handle the pipeline of anomaly detection using Isolation Forest
def anomaly_detection_pipeline(df, contamination=0.01):
    features = df.drop(columns=['Class'])  # Drop target column
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)

    # Initialize Isolation Forest model
    iso_forest = IsolationForest(contamination=contamination, random_state=42)
    preds = iso_forest.fit_predict(scaled_features)

    # Convert predictions (-1 -> anomaly, 1 -> normal) to binary (0 for fraud, 1 for non-fraud)
    return pd.Series(preds, name='predictions').map({-1: 1, 1: 0})

# Run the anomaly detection
df['predictions'] = anomaly_detection_pipeline(df)
```

Evaluating Performance (Accuracy, Precision, Recall)

```python
# Display the confusion matrix and classification report
cm = confusion_matrix(df['Class'], df['predictions'])

# Confusion Matrix
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# Classification Report
print(classification_report(df['Class'], df['predictions']))
```

ROC Curve and AUC Score

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(df['Class'], df['predictions'])

# Calculate AUC (Area Under Curve)
roc_auc = roc_auc_score(df['Class'], df['predictions'])
print(f"ROC-AUC Score: {roc_auc:.4f}")

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

Clustering Anomaly Detection Using DBSCAN

```python
from sklearn.cluster import DBSCAN

# Apply DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
dbscan = DBSCAN(eps=0.3, min_samples=10)
df['dbscan_preds'] = dbscan.fit_predict(scaled_features)

# Convert DBSCAN predictions to binary labels (1 for fraud, 0 for non-fraud)
df['dbscan_fraud'] = df['dbscan_preds'].apply(lambda x: 1 if x == -1 else 0)

# Display the results
print(df[['Class', 'dbscan_fraud']].head())
```

Visualizing PCA Results for Fraud vs Non-Fraud

```python
# Apply PCA to reduce the dimensionality for visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)

# Add PCA components to DataFrame for visualization
df_pca = pd.DataFrame(pca_components, columns=['PCA1', 'PCA2'])
df_pca['Class'] = df['Class']

# Visualize the PCA results
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Class', data=df_pca, palette='coolwarm', alpha=0.6)
plt.title('PCA Visualization: Fraud vs Non-Fraud')
plt.show()
```

Anomaly Detection Using One-Class SVM

```python
from sklearn.svm import OneClassSVM

# Apply One-Class SVM for anomaly detection
ocsvm = OneClassSVM(kernel='rbf', nu=0.01, gamma='auto')
df['ocsvm_preds'] = ocsvm.fit_predict(scaled_features)

# Convert predictions to binary (1 for fraud, 0 for non-fraud)
df['ocsvm_fraud'] = df['ocsvm_preds'].apply(lambda x: 1 if x == -1 else 0)

# Display the results
print(df[['Class', 'ocsvm_fraud']].head())
```

Comparing Multiple Models


```python
# Compare Isolation Forest, DBSCAN, and One-Class SVM
models = ['Isolation Forest', 'DBSCAN', 'One-Class SVM']
results = {
    'Model': models,
    'ROC-AUC': [roc_auc_score(df['Class'], df['predictions']),
                roc_auc_score(df['Class'], df['dbscan_fraud']),
                roc_auc_score(df['Class'], df['ocsvm_fraud'])]
}

comparison_df = pd.DataFrame(results)

# Plot the comparison
plt.figure(figsize=(8, 6))
sns.barplot(x='Model', y='ROC-AUC', data=comparison_df, palette='viridis')
plt.title('Model Comparison (ROC-AUC Score)')
plt.show()
```

Feature Importance with Random Forest

Time Series Analysis for Transaction Patterns

Removing Noise Using a Low Pass Filter

```python
# Apply a low-pass filter to smooth out data noise in the 'Amount' feature
from scipy.signal import butter, filtfilt

# Define low-pass filter function
def low_pass_filter(data, cutoff=0.1, fs=1.0):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(1, normal_cutoff, btype='low', analog=False)
    return filtfilt(b, a, data)

# Apply low-pass filter to the 'Amount' feature
df['Amount_filtered'] = low_pass_filter(df['Amount'])

# Visualize the original vs filtered 'Amount'
plt.figure(figsize=(12, 6))
plt.plot(df['Amount'], label='Original Amount', alpha=0.7)
plt.plot(df['Amount_filtered'], label='Filtered Amount', linestyle='--', color='red')
plt.title('Original vs Filtered Transaction Amount')
plt.legend()
plt.show()
```

Generating Synthetic Data Using SMOTE

```python
# Use SMOTE (Synthetic Minority Over-sampling Technique) to generate synthetic data for the minority class (fraudulent transactions)
from imblearn.over_sampling import SMOTE

# Apply SMOTE to the data (only on the features, not the target)
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(scaled_features, df['Class'])

# Display the new class distribution
print(f"New Class Distribution after SMOTE:\n{pd.Series(y_resampled).value_counts()}")
```
